```python
def find_minimal_covering_subsets(A, Q):
    result = []  # List to store the chosen subsets
    remaining_numbers = set(Q)  # Set of numbers not covered yet

    while remaining_numbers:
        best_subset = max(A, key=lambda subset: len(subset.intersection(remaining_numbers)))

        result.append(best_subset)
        remaining_numbers -= best_subset

    return result

# Example usage:
A = [{1, 2, 3, 4}, {2, 3, 5}, {4, 5, 6}]
Q = {2, 4, 6}

min_covering_subsets = find_minimal_covering_subsets(A, Q)
print(min_covering_subsets)
```
